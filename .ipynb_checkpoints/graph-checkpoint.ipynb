```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("data/august2019.tsv", sep="\t")
df.head()
```

```python
fig, (ax1, ax2, ax3, ax4) = plt.subplots(4,1, sharex=True)
fig.tight_layout()
fig.suptitle('August 2019')

ax1.plot(df['LOCATION'], df['TURBIDITY (NTU)'], 'r')
ax1.set_title('Turbidity')

ax2.plot(df['LOCATION'], df['PH VALUE'], 'b')
ax2.set_title('Ph Value')


ax3.plot(df['LOCATION'], df['EC VALUE'], 'g')
ax3.set_title('Electrical Conductivity')

ax4.plot(df['LOCATION'], df['ECOLI NO'],  'k')
ax4.set_title('Ecoli Count')
ax4.tick_params(labelrotation=90)
ax1.grid()
ax2.grid()
ax3.grid()
ax4.grid()

plt.show()

```

```python
corr = df[[ 'PH VALUE', 'EC VALUE', 'ECOLI NO']]
corr.corr(method='pearson') # {pearson, kendall, spearman} 
```

```python
fig.savefig('exports/august2019.png')
```
